# Introduction: Data Science Analysis on Ames Housing Data

===============================================
##### Problem
===============================================

Housing prices have mostly been transacted, over the counter, where buyer and seller both agrees that should be the market price, based on their indivudal interpretation.
Due to market inefficiencies, such as market knowledge and availbility of pricing tools, sellers might have been selling under the market price, and buyers might have been over paying the market price.

But what is market price ? No one can accurately say what the market price is down to the dollar and cents.

Our team seeks to find the market price for the next property being listed on the market. Through analysing characteristics of the house and historical transacted prices, we study the relationship and formulate a model/equation to predict to the market price. 


===============================================
##### Solution
===============================================

With such a pricing model, it will benefit the chain of participants in the real estate industry, from developers to retail clients, because with a price efficient model, it will facilitate more transacations and smoother transactions, as participants cannot be contesting the price model, unless they have empirical results and analysis done. 

For this pricing model, we will use Linear Regression model, and regularization techniques, such as Lasso and Ridge penatly, if there is an over or underfit in the model.

We will be using Ames Housing Dataset, which contains details of the house that has been transacted, such as, price sold, sq feet, pool quality, basement quality etc. These are qualitative and quantitaive details of the house and there are up to 70 of such qualities collected in the dataset. 


# Content Overview

Work flow / Process Flow 

- Import / Clean historical data 
- EDA : to determine which factors to analyse, via EDA we can see what factors to analyse 
- Build a model on the factors/features, with a train/test split approach 
- Enhanced that model with Linear Regression and regularization techniques Lasso and Ridge penalty. 
- Testing that model on the test portion of historical data, and test portion of historical data
- Evaluate that model results, via R Squared and MSE

```python
# split the dataset that has lot frontage and missing lot frontage
df_missing_lot_frontage = df[df['Lot Frontage'].isnull()]
print (df_missing_lot_frontage.shape)

df_non_missing_lot_frontage = df[df['Lot Frontage'] >= 0]
print (df_non_missing_lot_frontage.shape)
```

```python
# Using lot area as predictor of lot frontage
X = df_non_missing_lot_frontage[['Lot Area']]
y = df_non_missing_lot_frontage['Lot Frontage']


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1988)
lr = LinearRegression()
lr.fit(X_train,y_train)

print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

```python
# use 1st flr sf to predict lot frontage
X = df_non_missing_lot_frontage[['1st Flr SF']]
y = df_non_missing_lot_frontage['Lot Frontage']


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1988)
lr = LinearRegression()
lr.fit(X_train,y_train)

print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

# Mapping Ordinal Features

The ordinal dict will contain features that can be ranked as discrete variables. 

An example would be Lot Shape 

A house would have a Lot Shape, and it is considered IR3 , IR2 , IR1 , and Reg. 
These values will be mapped as 3 , 2 ,1 ,0 respectively, because IR3 is considered as Irregular, and IR2 is moderately irregular, hence the different degrees of regularity will be mapped from the least irregular to the most irregular, from 0 to 3. 

Assumption has been made that being the least irregular should be a 0 , and not a 3, as it could be vice versa.

# Modelling with Regularization, Lasso & Ridge

- Regularization is designed to penalize model complexity, therefore the higher the alpha, the less complex the model, decreasing the error due to variance (overfit). Alphas that are too high on the other hand increase the error due to bias (underfit). It is important, therefore to choose an optimal alpha such that the error is minimized in both directions.

```python
ss = StandardScaler()
Xs = ss.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(Xs, y, random_state=1988)
```

### Ridge optimization & Score Result

```python
#finding optimal alpha
ridge_alphas = np.logspace(0, 5, 200)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=5)
optimal_ridge.fit(Xs, y)

print (optimal_ridge.alpha_)
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = -cross_val_score(ridge, Xs, y, cv=5, scoring ='neg_root_mean_squared_error')

print (ridge_scores)
print (ridge_scores.mean())
```

### Lasso optimization & Score Result

```python
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(Xs, y)
```

```python
print(optimal_lasso.alpha_)
```

As alpha is 453.119, it is the most optimal, as it balances the model from being overfit or underfit. 
As from the graph, if alpha increases on the axis, the error of the score will increase upwards, i.e. a negative impact on the scoring of the model. This will result to the model from failure, i.e. just using a baseline model will surpass the need for a linear regression model with lasso regularisation. 

```python
#model5 = LassoCV(alphas=optimal_lasso.alpha_)
visualizer5 = AlphaSelection(optimal_lasso)
visualizer5.fit(Xs, y)
visualizer5.show()
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = -cross_val_score(lasso, Xs, y, cv=5,scoring ='neg_root_mean_squared_error')

print(lasso_scores.mean())
```

## Limitations
- Hyperparameter tuning: A predefined range of values was passed in to tune the hyperparameters for our regularisation models. There is a possibility that there are values outside of the predefined range that could have produced better results.

- Other external factors not taken into account: The value of a house is not purely determined by the features we have analysed. There are many other factors which could affect house demand and supply, such as the current economic climate, housing policies, demographic changes, proximity of the house to amenities like malls and schools, and so on. Predictions from our model are limited in its accuracy as it does not consider these factors.

- Limited timeframe: Our data comprised only transactions between 20062010. This is a pretty short timeframe, which makes it hard to capture actual general trends in sale prices of houses in Ames. The housing market within the 2006-2010 timeframe could be very different from how it is in present time.
Missing values: There were numerous missing values for which we did a convenient imputation for. This has definitely introduced a certain level of inaccuracy into our analysis.

- Limited generalisability to houses outside of Ames, Iowa: The predictive model we've built only applies to the houses in Ames and will not be accurate when applied to houses in other cities or countries. To make it more applicable across the board, we could perhaps remove features specific to houses in the U.S. or Ames, such as replacing Ames' neighborhood with a metric describing a property's distance from the city centre, or whether a property is near a buzzing commercial hub, etc.
