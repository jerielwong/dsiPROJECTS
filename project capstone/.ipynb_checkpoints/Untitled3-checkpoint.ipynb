# Linear Regression Regularisation 

 ### Linear Regression - Ridge / R2 
 - As alpha increases , the coefficients decays closes to 0 , but no 0 

```python
# RidgeCV what it does is it finds the optimal alpha, and then fits into the model 
# alphas is being assigned to 200 generated numbers from 10^1 to 10^5  <-- logspace 0,5,200 means this
# if it is linspace 0,5, 200 it is from value 0 to 5 with 200 points
# ridge will automatically find the best alpha here 
# .alpha_ is the optimal alpha


ridge = RidgeCV(alphas=(np.logspace(0,5,200)) ,cv=10)
ridge.fit(Xs_train,y_train)

```

```python
results['lr_ridge_pred'] = ridge.predict(Xs_test)
results['lr_ridge_pred'] = round(results['lr_ridge_pred'] ,2)
results
```

```python
metric = dict()
metric['Mean Absolute Error'] = mean_absolute_error(y_test, ridge.predict(Xs_test),multioutput='uniform_average')
metric['R2 Train Score'] = ridge.score(Xs_train,y_train)
metric['R2 Test Score'] = ridge.score(Xs_test,y_test)
metrics['lr_ridge'] = metric
pd.DataFrame(metrics)
```

```python
coefficients = pd.DataFrame({'variable':X_train.columns,'ridge_coef':ridge.coef_})
coefficients.sort_values('ridge_coef', inplace=True, ascending=False)
coefficients
```

# Random Forrest / Decision Tree  --- NO SCALING ---


```python
# Building GridSearch pipe 
pipe = Pipeline([ 
                ('dt',DecisionTreeRegressor(random_state = 42))
])


params = {
    'dt__max_depth':[2,3,5,7],
    'dt__min_samples_split': [5, 10, 15, 20],
    'dt__min_samples_leaf': [2, 3, 4, 5, 6],
    'dt__ccp_alpha': [0, 0.001, 0.01, 0.1, 1, 10]
}


grid = GridSearchCV(estimator = pipe, param_grid = params, cv =5)

grid.fit(X_train,y_train )

```

```python
print(grid.best_params_)
print(grid.score(X_train,y_train))
print(grid.score(X_test,y_test))
```

```python
results['rf_pred'] = grid.predict(X_test)
results
```

# Deep Learning - Neural Networks 

```python
def model_func(layer_one_neurons=100, layer_two_neurons=100):
    model = Sequential()
    
    model.add(Dense(layer_one_neurons, activation='relu', input_shape=(Xs_train.shape[1],)))
    
    model.add(Dense(layer_two_neurons, activation='relu'))
    
    model.add(Dense(1, activation=None))
    
    model.compile(loss='mse', optimizer='adam')
    
    return model
```

```python
# Modell Instantiation
nn = KerasRegressor(build_fn=model_func, batch_size=512, verbose=0)


# Pipe 
pipe = Pipeline([
    ('nn',nn)
])

# Params
params = {
    'nn__epochs': [10,100]
}

# Grid Searching
gs = GridSearchCV(pipe, param_grid=params, cv=2)


gs.fit(Xs_train, y_train)

print(gs.best_score_)
print(gs.best_params_)
```

```python
print(gs.get_params)
```

## Regularization - Deep Learning - L2 / Ridge Regression

Alpha used here is 0.01 


```python
model_l2 = Sequential()
model_l2.add(Dense(100, 
                   input_shape=(Xs_train.shape[1],),
                   activation='relu',
                   kernel_regularizer = regularizers.l2(0.01))) #1st layer
model_l2.add(Dense(100,
                   activation='relu',
                   kernel_regularizer =regularizers.l2(0.01))) #2nd layer 
model_l2.add(Dense(1))
model_l2.compile(loss='mse', optimizer='adam', metrics=['mae'])
history_l2= model_l2.fit(Xs_train, y_train, validation_data=(Xs_test, y_test), epochs=100, batch_size=512, verbose=2)
```

```python
train_loss_l2 = history_l2.history['loss']
test_loss_l2 = history_l2.history['val_loss']
plt.figure(figsize=(12, 8))
plt.plot(train_loss, label='Training loss', color='navy')
plt.plot(test_loss, label='Testing loss', color='skyblue')
plt.plot(train_loss_l2, label='L2 Training loss', color='darkred')
plt.plot(test_loss_l2, label='L2 Testing loss', color='pink')
plt.legend();
```

## Regularization - Deep Learning - Dropout Regularization


```python
model_dropout = Sequential()
model_dropout.add(Dense(100,
                        input_shape=(Xs_train.shape[1],),
                        activation='relu'
                       ))
model_dropout.add(Dropout(0.5))
model_dropout.add(Dense(100,
                        activation='relu'))

model_dropout.add(Dense(1))
model_dropout.compile(loss='mse', optimizer='adam', metrics=['mae'])
history_dropout = model_dropout.fit(Xs_train, y_train, validation_data=(Xs_test, y_test), 
                                    epochs=100, batch_size=512, verbose=2)
```

```python
train_loss_dropout = history_dropout.history['loss']
test_loss_dropout = history_dropout.history['val_loss']
plt.figure(figsize=(12, 8))
plt.plot(train_loss, label='Training loss', color='navy')
plt.plot(test_loss, label='Testing loss', color='skyblue')
plt.plot(train_loss_dropout, label='Dropout Training loss', color='darkgreen')
plt.plot(test_loss_dropout, label='Dropout Testing loss', color='lightgreen')
plt.legend();
```

## Regularization - Deep Learning - Early Stopping

```python
model_es = Sequential()
model_es.add(Dense(100,
                    input_shape=(Xs_train.shape[1],),
                    activation='relu'
                       ))
model_es.add(Dense(100,activation='relu'))
model_es.add(Dense(1))

model_es.compile(loss='mse', optimizer='adam', metrics=['mae'])
early_stop = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, mode='auto')
history_es = model_es.fit(Xs_train, y_train, validation_data=(Xs_test, y_test), 
                          epochs=100, batch_size=None, verbose=2,
                         callbacks=[early_stop])
```

```python
train_loss_es = history_es.history['loss']
test_loss_es = history_es.history['val_loss']
plt.figure(figsize=(12, 8))
plt.plot(train_loss, label='Training loss', color='yellow')
plt.plot(test_loss, label='Testing loss', color='green')
plt.plot(train_loss_es, label='ES Training loss', color='navy')
plt.plot(test_loss_es, label='ES Testing loss', color='red')
plt.legend();
```
