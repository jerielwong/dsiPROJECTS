```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df = pd.read_csv('./dataset/25 Sep 2020 finviz.csv')
```

```python
df.loc[df['Ticker'] == 'AAPL' ]
```

```python
#function to remove %
def pct_remove(value1):
    if '%' in str(value1):
        value1 = value1.replace('%','')
    return value1
```

```python
    df['50-Day Simple Moving Average']=df['50-Day Simple Moving Average'].map(pct_remove)
    df['50-Day Simple Moving Average']=df['50-Day Simple Moving Average'].astype('float64')
    df['50-Day Simple Moving Average']=df['50-Day Simple Moving Average']/100
```

```python
    df['200-Day Simple Moving Average']=df['200-Day Simple Moving Average'].map(pct_remove)
    df['200-Day Simple Moving Average']=df['200-Day Simple Moving Average'].astype('float64')
    df['200-Day Simple Moving Average']=df['200-Day Simple Moving Average']/100
```

```python
    df['Change']=df['Change'].map(pct_remove)
    df['Change']=df['Change'].astype('float64')
    df['Change']=df['Change']/100
```

```python
df.loc[df['Ticker'] == 'AAPL' ]
```

```python
df['50 sma price'] = (df['Price']) / (1 +df['50-Day Simple Moving Average'] )
```

```python
df['prev close'] = (df['Price']) / (1 + df['Change'] )
df['change value'] = df['prev close'] * df['Change']
```
